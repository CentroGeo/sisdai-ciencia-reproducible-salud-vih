```python
import pandas as pd
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

## **Estadísticas generales (visualizaciones de barras)** 


```python
### Obtener datos 
nombre_archivo = os.path.join('..', 'datos', 'datos_procesados', 'nodos.csv')
datos = pd.read_csv(nombre_archivo)

### Reemplazar valores nulos 
for column in datos:
        
    datos[column] =datos[column].replace(np.nan, '')  
    datos = datos[(datos['cv'] != '') & (datos['cd4_a'] != '') & (datos['edad'] != '')]

### Calcular promedio de Linfocitos con respecto a edad y carga viral con respecto a edad (por rango)
datos_aux = datos[['edad_rango','cv', 'cd4_a']]
datos_aux = datos_aux.replace('',np.nan).dropna()
dfb = datos_aux.groupby(['edad_rango']).agg({'cv':'mean','cd4_a':'mean'}).round(decimals=2).reset_index()

### Distribución de carga viral y distribución de linfocitos
dc = datos[['cv']]
dd = datos[['cd4_a']]
```

```python
# Crear base de datos de distribución de carga viral
minor0_5 = []
minor1 = []
minor1_5 = []
minor2 = []
minor2_5 = []
minor3 = []
minor3_5 = []
minor4 = []
minor4_5 = []
minor5 = []
minor5_5 = []
minor6 = []
minor6_5 = []
minor7 = []

dc_cv = dc['cv'].to_list()

for i in range(len(dc_cv)):
    if(dc_cv[i] >= 0 and dc_cv[i] <= 0.5):
        minor0_5.append(dc_cv[i])
    if(dc_cv[i] > 0.5 and dc_cv[i] <= 1):
        minor1.append(dc_cv[i])
    if(dc_cv[i] > 1 and dc_cv[i] <= 1.5):
        minor1_5.append(dc_cv[i])
    if(dc_cv[i] > 1.5 and dc_cv[i] <= 2):
        minor2.append(dc_cv[i])
    if(dc_cv[i] > 2 and dc_cv[i] <= 2.5):
        minor2_5.append(dc_cv[i])
    if(dc_cv[i] > 2.5 and dc_cv[i] <= 3):
        minor3.append(dc_cv[i])
    if(dc_cv[i] > 3 and dc_cv[i] <= 3.5):
        minor3_5.append(dc_cv[i])
    if(dc_cv[i] > 3.5 and dc_cv[i] <= 4):
        minor4.append(dc_cv[i])
    if(dc_cv[i] > 4 and dc_cv[i] <= 4.5):
        minor4_5.append(dc_cv[i])
    if(dc_cv[i] > 4.5 and dc_cv[i] <= 5):
        minor5.append(dc_cv[i])
    if(dc_cv[i] > 5 and dc_cv[i] <= 5.5):
        minor5_5.append(dc_cv[i])
    if(dc_cv[i] > 5.5 and dc_cv[i] <= 6):
        minor6.append(dc_cv[i])
    if(dc_cv[i] > 6 and dc_cv[i] <= 6.5):
        minor6_5.append(dc_cv[i])
    if(dc_cv[i] > 6.5 and dc_cv[i] <= 7):
        minor7.append(dc_cv[i])

d = {
    'nombre': [
    '0.5', '1','1.5', '2', '2.5', '3', '3.5', '4', '4.5', '5', '5.5', '6', '6.5', '7'
    ],
    'cantidad': [
        len(minor0_5), len(minor1), len(minor1_5), len(minor2), len(minor2_5), len(minor3), len(minor3_5), 
        len(minor4), len(minor4_5), len(minor5), len(minor5_5), len(minor6), len(minor6_5), len(minor7)
    ]
    }
dt = pd.DataFrame(data=d)
```

```python
# Crear base de datos de distribución de linfocitos

minor100 = []
minor200 = []
minor300 = []
minor400 = []
minor500 = []
minor600 = []
minor700 = []
minor800 = []
minor900 = []
minor1000 = []
minor1100 = []
minor1200 = []
minor1300 = []
minor1400 = []

dd_cd4a = dd['cd4_a'].to_list()

for i in range(len(dd_cd4a)):
    if(dd_cd4a[i] >= 0 and dd_cd4a[i] <= 100):
        minor100.append(dd_cd4a[i])
    if(dd_cd4a[i] > 100 and dd_cd4a[i] <= 200):
        minor200.append(dd_cd4a[i])
    if(dd_cd4a[i] > 200 and dd_cd4a[i] <= 300):
        minor300.append(dd_cd4a[i])
    if(dd_cd4a[i] > 300 and dd_cd4a[i] <= 400):
        minor400.append(dd_cd4a[i])
    if(dd_cd4a[i] > 400 and dd_cd4a[i] <= 500):
        minor500.append(dd_cd4a[i])
    if(dd_cd4a[i] > 500 and dd_cd4a[i] <= 600):
        minor600.append(dd_cd4a[i])
    if(dd_cd4a[i] > 600 and dd_cd4a[i] <= 700):
        minor700.append(dd_cd4a[i])
    if(dd_cd4a[i] > 700 and dd_cd4a[i] <= 800):
        minor800.append(dd_cd4a[i])
    if(dd_cd4a[i] > 800 and dd_cd4a[i] <= 900):
        minor900.append(dd_cd4a[i])
    if(dd_cd4a[i] > 900 and dd_cd4a[i] <= 1000):
        minor1000.append(dd_cd4a[i])
    if(dd_cd4a[i] > 1000 and dd_cd4a[i] <= 1100):
        minor1100.append(dd_cd4a[i])
    if(dd_cd4a[i] > 1100 and dd_cd4a[i] <= 1200):
        minor1200.append(dd_cd4a[i])
    if(dd_cd4a[i] > 1200 and dd_cd4a[i] <= 1300):
        minor1300.append(dd_cd4a[i])
    if(dd_cd4a[i] > 1300 and dd_cd4a[i] <= 1400):
        minor1400.append(dd_cd4a[i])

s = {
    'cd4_a': [
    '100', '200','300', '400', '500', '600', '700', '800', '900', '1000', '1100', '1200', '1300', '1400'
    ],
    'cantidad': [
        len(minor100), len(minor200), len(minor300), len(minor400), len(minor500), len(minor600), len(minor700), 
        len(minor800), len(minor900), len(minor1000), len(minor1100), len(minor1200), len(minor1300), len(minor1400)
    ]
    }
du = pd.DataFrame(data=s)
du = du.rename({'cd4_a': 'linfocitos'}, axis=1) 
```
